```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC") ## Importing data
```

```python
sp500 = sp500.history(period="max") ## Importing data
```

```python
sp500.plot.line(y="Close", use_index=True) # Y = closing price # X = date
```

```python
del sp500["Dividends"] ## dont need these
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1) # shifting date back one gives us the hypothetical "tomorrow" to predict
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int) # target == 1 if the price went up from the previous day and 0 if otherwise.
```
